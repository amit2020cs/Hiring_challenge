There are total 37 missing data-point in 12 in C1 and C2 each and 13 in C14.

```python
df.fillna(df.mean(), inplace=True)
```

Replace the missing numerical data with mean of their column.

```python
df  = df.replace('?', np.NaN)
```

```python
df.isnull().sum()
```

```python
# fill null datapoint with most occur in that coloumn.
for col in df.columns:
    if df[col].dtype == 'object':
        df = df.fillna(df[col].value_counts().index[0])
```

```python
df.isnull().sum()
```

```python
# Preprocess data using LabelEncoder to change all categorical data into numerical data.
from sklearn.preprocessing import LabelEncoder
```

```python
# Initialise LabelEncoder
label_encoder = LabelEncoder()
```

```python
# Apply to all the column having dtype:Object
for col in df.columns.values:
    if df[col].dtype=='object':
        df[col]=label_encoder.fit_transform(df[col])
```
